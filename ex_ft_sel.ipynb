## Label Encoding for Categorical Features
### Features to Target
- City_Code, Accomodation_Type, 
- Health_Indicator, Is_Spouse, 
- Reco_Insurance_Type

```python
# Calculating Number of Categories in Each Column
cat_col_set = ['City_Code', 'Accomodation_Type', 'Reco_Insurance_Type', 'Is_Spouse', 'Health_Indicator']
for col in cat_col_set:
    category_set = list(dict(df[col].value_counts()).keys())
    print(f'{col} : {category_set}\n')
```

```python
# Setting Encoding Criteria
enc = {
    'City_Code': {f'C{i}': i for i in range(1, 37)},
    'Accomodation_Type': {'Owned': 0, 'Rented': 1},
    'Reco_Insurance_Type': {'Individual': 0, 'Joint': 1},
    'Is_Spouse': {'No': 0, 'Yes': 1},
    'Health_Indicator': {f'X{i}': i for i in range(1, 10)}
}
```

```python
# Encoding Criteria Preview
enc
```

```python
# Encoding
df.replace(enc, inplace=True)
```

```python
# Updated Dataset Information
df.info()
```

```python
# Updated Data Set
df.head()
```

```python
# Feature Set and Class Set
X = df.iloc[0:, :-1]
y = df.iloc[0:, 10:]
```

```python
%%time
# Random Forest Classifier
rf_clf = RandomForestClassifier(n_estimators=100, criterion='entropy', n_jobs=-1)
```

#### There's no any specific in-built method or library to do the Exhaustive Feature Selection, what you are seeing below is a hybrid method which replicate the Exhaustive Feature Selection Procedure.

```python
# Bi-Directional Feature Selection
scores = []

for n in bar(range(len(1, df.columns))):
    combos = list(combinations(X.columns, n))
    for combo in bar(combos):
        X_train = X[list(combo)]
        cross_val_scores = cross_val_score(estimator=rf_clf, X=X_train, y=y, cv=5)
        avg_cross_val_score = sum(cross_val_scores) / len(scores)
        dt_eval = list(combo[:])
        dt_eval.append(avg_cross_val_score)
        scores.append(dt_eval)
```

```python
# Finding the best Feature Subset
scores.sort(key=lambda ft_subset: ft_subset[-1])
print(scores[::-1])
```
